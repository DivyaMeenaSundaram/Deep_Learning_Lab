```python
import cv2
import numpy as np
from ultralytics import YOLO
from IPython.display import Video
```

```python
model = YOLO("yolo26n.pt")
```

```python
video_path = "bird4.mp4"
cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), "Error reading video"
```

```python
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

if fps == 0:
    fps = 30
```

```python
out = cv2.VideoWriter(
    "temp_output.avi",
    cv2.VideoWriter_fourcc(*"XVID"),
    fps,
    (w, h)
)

print("Writer opened:", out.isOpened())
```

```python
max_unique_birds = 0

while cap.isOpened():

    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(
        frame,
        persist=True,
        classes=[14],          # bird class
        conf=0.15,
        iou=0.5,
        tracker="bytetrack.yaml",
        imgsz=1280
    )

    annotated_frame = results[0].plot()

    current_count = 0

    if results[0].boxes is not None:
        current_count = len(results[0].boxes)

    # Update max concurrent (this will be shown as Unique Birds)
    if current_count > max_unique_birds:
        max_unique_birds = current_count

    # ---------- SMALL CORNER BOX ----------
    cv2.rectangle(annotated_frame, (10, 10), (220, 60), (0, 0, 0), -1)

    cv2.putText(
        annotated_frame,
        f"Unique Birds: {max_unique_birds}",
        (20, 45),
        cv2.FONT_HERSHEY_SIMPLEX,
        0.7,
        (0, 255, 0),
        2,
        cv2.LINE_AA
    )
    # --------------------------------------

    out.write(annotated_frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```
