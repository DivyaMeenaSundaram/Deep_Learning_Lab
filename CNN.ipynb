```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 1️⃣ Load an Image (Change "cat.jpg" to your file)
image = cv2.imread("/content/cat1.jpg", cv2.IMREAD_GRAYSCALE)  # Convert to grayscale

# Resize for better visualization
image = cv2.resize(image, (300, 300))

# 2️⃣ Define a Filter (Edge Detection Kernel)
kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

# 3️⃣ Perform Convolution
output = convolve2d(image, kernel, mode="same", boundary="fill", fillvalue=0)

# 4️⃣ Display the Original and Convolved Image
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original Image
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis("off")

# Convolved Image (Edge Detected)
axes[1].imshow(output, cmap='gray')
axes[1].set_title("Filtered Image (Edge Detection)")
axes[1].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 1️⃣ Load and Preprocess the Image
image = cv2.imread("/content/cat1.jpg", cv2.IMREAD_GRAYSCALE)  # Convert to grayscale
image = cv2.resize(image, (300, 300))  # Resize for consistency

# 2️⃣ Define Convolution Filters
# Sobel Edge Detection (Detects Vertical Edges)
sobel_kernel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

# Gaussian Blur Kernel (5x5 Approximation)
gaussian_kernel = (1/16) * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
])

# Sharpening Kernel
sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

# 3️⃣ Apply Convolutions
edge_detected = convolve2d(image, sobel_kernel, mode="same", boundary="fill", fillvalue=0)
blurred = convolve2d(image, gaussian_kernel, mode="same", boundary="fill", fillvalue=0)
sharpened = convolve2d(image, sharpen_kernel, mode="same", boundary="fill", fillvalue=0)

# 4️⃣ Display Results
fig, axes = plt.subplots(1, 4, figsize=(16, 6))

# Original Image
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis("off")

# Edge Detection (Sobel)
axes[1].imshow(edge_detected, cmap='gray')
axes[1].set_title("Edge Detection (Sobel)")
axes[1].axis("off")

# Gaussian Blur
axes[2].imshow(blurred, cmap='gray')
axes[2].set_title("Gaussian Blur")
axes[2].axis("off")

# Sharpened Image
axes[3].imshow(sharpened, cmap='gray')
axes[3].set_title("Sharpened Image")
axes[3].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# 1️⃣ Load and Preprocess the Image
image = cv2.imread("/content/cat1.jpg")  # Load color image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR (OpenCV format) to RGB
image = cv2.resize(image, (300, 300))  # Resize for consistency

# Function to Apply Convolution to Each Channel (R, G, B)
def apply_convolution(image, kernel):
    channels = []
    for i in range(3):  # Loop through R, G, B channels
        convolved = convolve2d(image[:, :, i], kernel, mode="same", boundary="fill", fillvalue=0)
        channels.append(convolved)
    return np.stack(channels, axis=-1)  # Merge channels back into an image

# 2️⃣ Define Convolution Filters
# Sobel Edge Detection (Detects Vertical Edges)
sobel_kernel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

# Gaussian Blur Kernel (Smoothing)
gaussian_kernel = (1/16) * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
])

# Sharpening Kernel
sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

# 3️⃣ Apply Filters to RGB Image
edge_detected = apply_convolution(image, sobel_kernel).astype(np.uint8)
blurred = apply_convolution(image, gaussian_kernel).astype(np.uint8)
sharpened = apply_convolution(image, sharpen_kernel).astype(np.uint8)

# 4️⃣ Display Results
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Original Image
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

# Edge Detection
axes[1].imshow(edge_detected)
axes[1].set_title("Edge Detection (Sobel)")
axes[1].axis("off")

# Gaussian Blur
axes[2].imshow(blurred)
axes[2].set_title("Gaussian Blur")
axes[2].axis("off")

# Sharpened Image
axes[3].imshow(sharpened)
axes[3].set_title("Sharpened Image")
axes[3].axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1️⃣ Load and Preprocess the Image
image = cv2.imread("/content/cat1.jpg")  # Load color image
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
image = cv2.resize(image, (300, 300))  # Resize for consistency

# 2️⃣ Define Filters (Kernels)
# Sobel Edge Detection
sobel_kernel = np.array([
    [1, 0, -1],
    [2, 0, -2],
    [1, 0, -1]
])

# Gaussian Blur (5x5)
gaussian_kernel = (1/16) * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]
])

# Sharpening Kernel
sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

# 3️⃣ Apply Filters Using OpenCV
edge_detected = cv2.filter2D(image, -1, sobel_kernel)  # Edge detection
blurred = cv2.filter2D(image, -1, gaussian_kernel)  # Gaussian blur
sharpened = cv2.filter2D(image, -1, sharpen_kernel)  # Sharpening

# 4️⃣ Display Results
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

# Original Image
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[0].axis("off")

# Edge Detection
axes[1].imshow(edge_detected)
axes[1].set_title("Edge Detection (Sobel)")
axes[1].axis("off")

# Gaussian Blur
axes[2].imshow(blurred)
axes[2].set_title("Gaussian Blur")
axes[2].axis("off")

# Sharpened Image
axes[3].imshow(sharpened)
axes[3].set_title("Sharpened Image")
axes[3].axis("off")

plt.show()
```
